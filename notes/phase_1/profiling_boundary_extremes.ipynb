```python
import pandas as pd

# Contains profiling metrics for each note
profile_df = pd.read_csv("../../data/sample/profiling_per_note.csv")
# Contains raw text notes corresponding to the profiling data, indexed the same way
sample_df = pd.read_csv("../../data/sample/profiling_sample_500.csv")

# Define the lists of metrics to analyse
functional_metrics = [
    "colon_header_count",
    "uppercase_header_count",
    "numeric_token_count",
    "bp_pattern_count",
    "deid_token_count",
]

structural_metrics = [
    "char_length",
    "token_count",
    "line_count"
]

# ----------------------------
# 1. Collect functional metrics extremes
# ----------------------------

# Empty set to store unique indices of notes that are extreme in any metric
extreme_indices = set()

# Loop through each metric
for metric in functional_metrics:
    # Get top 5 and bottom 5 indices for the current metric
    top5 = profile_df.sort_values(metric, ascending=False).head(5).index
    bottom5 = profile_df.sort_values(metric).head(5).index
    
    # Add these indices to the set of extreme indices
    extreme_indices.update(top5)
    extreme_indices.update(bottom5)

print(f"Total unique extreme notes: {len(extreme_indices)}")

# ----------------------------
# 2. Collect structural metric extremes (char_length only)
# ----------------------------  

length_indices = set()

top2 = profile_df.sort_values("char_length", ascending=False).head(2).index
bottom2 = profile_df.sort_values("char_length").head(2).index

length_indices.update(top2)
length_indices.update(bottom2)

print(f"Total unique length-extreme notes: {len(length_indices)}")

all_extremes = extreme_indices.union(length_indices)
print(f"Total unique notes overall: {len(all_extremes)}")

# ----------------------------
# 3. Inspect each unique functional note
# ----------------------------

print("FUNCTIONAL EXTREMES:")

# Loop through each unique extreme index
for idx in extreme_indices:
    print("="*80)
    print(f"NOTE INDEX: {idx}")
    print(profile_df.loc[idx, functional_metrics])  # show all 5 functional metric values for the index
    print("-"*80)
    print(sample_df.loc[idx, "TEXT"])    # print raw note text
    print("\n\n")

# ----------------------------
# 4. Inspect each unique structural note
# ----------------------------

print("STRUCTURAL EXTREMES:")

for idx in length_indices:
    print("="*80)
    print(f"NOTE INDEX: {idx}")
    print(profile_df.loc[idx, structural_metrics])  # show all 3 structural metric values for the index
    print("-"*80)
    print(sample_df.loc[idx, "TEXT"])    # print raw note text
    print("\n\n")
```
